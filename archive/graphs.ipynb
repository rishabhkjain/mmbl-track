```python
import matplotlib.pyplot as plt
import numpy as np
import csv
%matplotlib inline
```

```python
frameLst = []
deltaLst = []

with open('12_detailedResult.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        if len(row) == 0:
            continue
        frameLst.append(int(row[0]))
        deltaLst.append(float(row[3]))
        

```

```python
print(frameLst)
```

```python
frameLst1 = frameLst[:10]
deltaLst1 = deltaLst[:10]
```

```python
print (frameLst1)
print (deltaLst1)
```

```python
totalLst = [0]*len(deltaLst)
for i in range (len(deltaLst)):
    if i == 0:
        totalLst[i] = deltaLst[i]
    else:
        
        totalLst[i] = deltaLst[i] + totalLst [i-1]
```

```python
print(totalLst)
```

```python
plt.plot(frameLst,totalLst, label='Displacement of Particle 12')
plt.axis([min(frameLst), max(frameLst), 0, max(totalLst) + 10])

plt.show()
```
